# Q 13 Which brands of vehicles are less affected by cost depreciation ?






```python


# Group by car name and calculate the average depreciation
average_depreciation_by_brand = data.groupby('Car_Name')['Depreciation'].mean()

# Sort by average depreciation in ascending order to find the least depreciated brands
least_depreciated_brands = average_depreciation_by_brand.sort_values(ascending=True)

print(f"Brands with the least depreciation:\n{least_depreciated_brands}")
```

# Q: 15 In general selling price is affected by age of vehicle and distance driven by vehicle , is it observable from data ?

```python
plt.scatter(data.Kms_Driven,data.Selling_Price)
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling_Price in lacs")
plt.title("Selling Price Vs Kilometers driven")
plt.show()
```

```python

# Calculate the correlation between selling price, age, and distance driven
data['Age'] = data['Year'].max() - data['Year']  # Calculate vehicle age
correlation_matrix = data[['Selling_Price', 'Age', 'Kms_Driven']].corr()
print(correlation_matrix)

# Visualize the relationships
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Selling Price, Age, and Distance Driven')
plt.show()

```

# Q: 20 Which is the most sold bike here?

```python


# Assuming vehicles with Selling_Price less than 3 lakh are bikes
bikes = data[data['Selling_Price'] < 3]

# Remove '800' and 'omni' cars from the list of bikes
filtered_bikes = bikes[~bikes['Car_Name'].isin(['800', 'omni'])]

# Find the most sold bike
most_sold_bike = filtered_bikes['Car_Name'].value_counts().idxmax()

print(f"The most sold bike is: {most_sold_bike}")
```
